Let's make some changes to the data

```python
df.head()
```

```python
# Making a new column
df["Total"] = df["HP"] + df["Attack"] + df["Defense"] + df["Sp. Atk"] + df["Sp. Def"] + df["Speed"]

df.head()
```

```python
df.dtypes["HP"]
```

```python
df = df.drop(columns = ["Total"])
df
```

```python
# Add column a different way

df["Total"] = df.iloc[:, 4:10].sum(axis = 1)   # axis = 1 means you want to add horizontally  
df
```

```python
# We want to move a column to a different spot

cols = list(df.columns)
df = df[cols[0:4] + [cols[-1]] + cols[4:12]]

df
```

Filtering Data

```python
# Instead of typing out 'and', we use & inside of a pandas dataframe
# Similarly, instead of saying 'or', we will use |

df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison')]
```

```python
df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison') & (df["HP"] > 70)]
```

```python
new_df = df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison') & (df["HP"] > 70)]

new_df
```

```python
# It can get annoying to have the old index numbers. In that case, we can rest them

new_df.reset_index(drop = True, inplace = True)  # the 'drop' is necessary in order to actually eliminate the old index #s

new_df
```

Regex Filtering (Filtering based on textual patterns)

```python
df.loc[df['Name'].str.contains('Mega')]
```

```python
df.loc[~df['Name'].str.contains('Mega')]        # The ~ represents "not"
```

```python
#Regex

import re   # regular expressions

df.loc[df['Type 1'].str.contains('Fire|Grass', regex = True)]

#df.loc[df['Type 1'].str.contains('fire|grass', flags = re.I, regex = True)]   # the glags = ... tells Python to ignore the capitalization of the words
```

```python
df.loc[df['Name'].str.contains('^Pi[a-z]*', regex = True)]


# the ^ is needed because I want the names to start with a 'Pi'
# The * is needed to show 0 or more of the a through z letters can come after 'Pi'
```

Aggregate Statistics (with Groupby)

```python
df = pd.read_csv('modified_pokemon_data_1.csv')

df
```

```python
df.groupby(['Type 1']).mean().sort_values('Defense', ascending = False)
```

```python
df.groupby(['Type 1']).mean().sort_values('Attack', ascending = False)
```

```python
df.groupby(['Type 1']).count()
```

```python
df.groupby(['Type 1']).sum()
```

```python
# Remember that it doesn't make sense to always use mean, count, and sum when making an analysis of something
```

```python
# To simplify the .count() method

df['count'] = 1

#df

df.groupby(["Type 1"]).count()['count']
```

```python
df.groupby(['Type 1', 'Type 2']).count()['count']
```
